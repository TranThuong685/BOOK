```python
# Import thư viện requests và BeautifulSoup
from bs4 import BeautifulSoup
import requests

# Định nghĩa url của trang web bạn muốn cào dữ liệu
url = "https://h2tstore.vn/collections/ao-thun"

# Gửi yêu cầu GET đến url và lấy nội dung HTML
response = requests.get(url)
html = response.content

# Tạo một đối tượng BeautifulSoup từ nội dung HTML
soup = BeautifulSoup(html, "html.parser")
```

```python
links = soup.find_all("a", href=lambda x: x and x.startswith("/products/"))
urls = set()
```

```python
data = []
product_detail = []
```

```python
for link in links:
    # Tạo url đầy đủ của liên kết bằng cách ghép với url gốc của trang web
    urls.add("https://h2tstore.vn" + link['href'])
urls = list(urls)
urls
```

```python
for url in urls:
    # Gửi yêu cầu GET đến product_url và lấy nội dung HTML
    product_response = requests.get(url)
    product_html = product_response.content

    # Tạo một đối tượng BeautifulSoup từ nội dung HTML
    product_soup = BeautifulSoup(product_html, "html.parser")
    
    # Tìm thẻ <h1> có class là "product-single__title"
    # Đây là thẻ chứa tên sản phẩm
    name = product_soup.find("div", class_="nameProductStore").text
    # Tìm thẻ <span> có id là "ProductPrice"
    # Đây là thẻ chứa giá bán của sản phẩm
    price = product_soup.find("div", class_="priceProductStore").text

    # Tìm thẻ <div> có id là "product-description"
    # Đây là thẻ chứa mô tả của sản phẩm
    description = product_soup.find("meta", property="og:description")['content']

    # Tạo một từ điển để lưu thông tin của sản phẩm
    product_info = {
        "name": name,
        "price": price,
        "description": description,
        "category": 'áo thun'
    }

    # Thêm từ điển vào danh sách data
    data.append(product_info)
pd.DataFrame(data)
```

```python
index = 0
for url in urls:
    response = await session.get(url)
    await response.html.arender(timeout=500)
    html = response.html
    options = html.find('option[data-title]') # Lấy ra tất cả các thẻ option có thuộc tính data-title
    for option in options: # Duyệt qua từng thẻ option trong danh sách kết quả
        data_title = option.attrs['data-title'].split(' / ') # Lấy ra giá trị của thuộc tính data-title
        product_info = {
            "name": data[index]['name'],
            "size": data_title[0],
            "color": data_title[1],
            "quantity": 50,
        }
        product_detail.append(product_info)
    index += 1
```

```python
# Import các thư viện cần thiết
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import os
```

```python
# Khởi tạo webdriver
driver = webdriver.Chrome(executable_path="chromedriver.exe")
```

```python
# Mở trang web bạn muốn lấy ảnh
url = "https://h2tstore.vn/collections/ao-thun"
driver.get(url)

# Lấy mã HTML của trang web
html = driver.page_source

# Tạo một đối tượng BeautifulSoup để phân tích HTML
soup = BeautifulSoup(html, "html.parser")
```

```python
links = soup.find_all("a", href=lambda x: x and x.startswith("/products/"))
urls = set()
```

```python
for link in links:
    # Tạo url đầy đủ của liên kết bằng cách ghép với url gốc của trang web
    urls.add("https://h2tstore.vn" + link['href'])
urls = list(urls)
urls
```
